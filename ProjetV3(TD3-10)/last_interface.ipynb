```python
import ipywidgets as widgets
```

```python
from CorpusV2 import Corpus
from SearchEngineV2 import SearchEngine
```

```python
corpus = Corpus("test")
corpus.load("test")
SearchEngine = SearchEngine(corpus)
```

```python
output = widgets.Output()
def on_button_click(b):        
    with output:
        output.clear_output()
        if wMots.value == '' or wSlide.value==0:
            print("Veuillez entrer des mots et un nombre de documents à chercher")
        else:
            df = SearchEngine.search(wMots.value, wSlide.value)
            display(df)
```

```python
wTitre = widgets.Label(value='Moteur de recherche')
wMots = widgets.Text(value='', 
                     description='Mots clés', 
                     disabled=False)
wSlide = widgets.IntSlider(description="Nombre d'article à extraire", 
                           min=1,
                           max=50,
                           value=10,
                           style={'description_width': 'initial'},
                           layout=widgets.Layout(width='400px'))
wBouton = widgets.Button(description='Rechercher')
wBouton.on_click(on_button_click)

display(widgets.VBox([wTitre,wMots,wSlide,wBouton,output],layout=widgets.Layout(align_items='center')))
```
